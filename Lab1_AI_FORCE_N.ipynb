```python
#Gstion des manquants ou nombre de valeurs non disponibles NaN
titanic.isnull().sum() #ou titanic.isna.sum()
```

```python
#Coemme le nombre de données manquantes n'est ci importante losqu"on mois de données manquantes
#c'est pourquoi il est interessante d calculer le poucentage
prct_manquants=titanic.isna().sum()*100/titanic.shape[0]
round(prct_manquants.sort_values(),2)
```

```python
# On peut encoder de la facon les données manquantes
#others_mv=['9', '99', '999']
#pd.readcsv(....... , na_values=other_mv)
```

```python
#visualisation des données manquantes par variable
fig=plt.figure(figsize=(20,3))
plt.bar(x=prct_manquants.sort_values().index , height=prct_manquants.sort_values())
plt.axhline(y=15, color='r', linestyle = '-')#ligne à 15%
plt.axhline(y=30, color = 'k', linestyle = '-') #ligne à 30%
```

```python
# Les 03 types de donnéées Manquantes (detection)
```

```python
#CAS des MCAR
#Il faut d'abord construire une matrice des indicateurs
to_change= list(prct_manquants[prct_manquants>0].index)
to_change
```

```python
#Tableau des indicateurs manquants ou non
bool_titanic=1*titanic[to_change].isnull()
bool_titanic.tail()
```

```python
from scipy.stats import chi2_contingency

def cramers_V(var1, var2) :
   """Evalue par une normalisation du chi-deux entre 0 et 100%
   correlation entre 2 variable catégrielle"""
   crosstab = np.array( pd.crosstab( var1, var2, rownames=None, colnames=None))#tableau de continger
   stats = chi2_contingency(crosstab)#[0]#valeur de la statistique du test
   obs = np.sum(crosstab)#nombre d'observations
   mini = min(crosstab.shape)-1  #min(p,q)-1
   return((np.sqrt(stats[0]/(obs*mini))), stats[1]*100)
```
