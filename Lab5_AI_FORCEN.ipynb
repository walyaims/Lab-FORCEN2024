#Exemple 1 : Extraire des données d'une API COVID open source
Dans cet exemple, nous nous connecterions à une API COVID Open source uniquement pour extraire et analyser les informations json de manière personnalisée.

#1. Connectez-vous à une API
Dans un premier temps, nous devons nous connecter à une API et établir une connexion sécurisée comme indiqué ci-dessous :

Dans cet article, nous avons utilisé l' API COVID19-India pour récupérer les données des cas de la liste par état.

```python
import requests
import json
response_API = requests.get('https://api.covid19india.org/state_district_wise.json')
#print(response_API.status_code)
```

```python
response_API
```

Alors que nous extrayons les données d'une API, nous avons utilisé la get()fonction pour obtenir les informations de l'API.

#2. Obtenez les données de l'API

Après avoir établi une connexion saine avec l'API, la tâche suivante consiste à extraire les données de l'API. Regardez le code ci-dessous !

```python
data = response_API.text
```

```python
data
```

Cela requests.get(api_path).text nous aide à extraire les données de l'API mentionnée.

#3. Analysez les données au format JSON

Après avoir extrait les données, il est maintenant temps de convertir et de décoder les données au format JSON approprié, comme indiqué ci-dessous :

```python
json.loads(data)
```

La fonction json.loads() analyse les données au format JSON .

#4. Extrayez les données et imprimez-les

Le format JSON contient des données dans un format clé-valeur qui ressemble à un dict Python . Ainsi, nous pouvons extraire et imprimer les données en utilisant les valeurs clés comme indiqué :

```python
parse_json = json.loads(data)
```

```python
active_case =parse_json['Andaman and Nicobar Islands']['districtData']['South Andaman']['active']
```

#Exemple 2 : Extraire des données d'une API GMAIL Open Source
Maintenant, connectons-nous et extrayons les données de l'API GMAIL . Cette API représente la structure générique et les informations que nous pouvons récupérer de l'API.

```python
import requests
import json
response_API = requests.get('https://gmail.googleapis.com/$discovery/rest?version=v1')
#print(response_API.status_code)
data = response_API.text
parse_json = json.loads(data)
info = parse_json['description']
print("Info about API:\n", info)
key = parse_json['parameters']['key']['description']
print("\nDescription about the key:\n",key)
```
